```python
import csv
import sqlite3
import pandas as pd

conn = sqlite3.connect('sars.db')

sars = pd.read_csv("data/sars_2003_complete_dataset_clean.csv")

sars.to_sql("sars", conn)
```

```python
% sql SELECT * from sars
```

```python
#MUST HAVE VIEW OR IT WONT ALTER ORIGINAL DATABASE

%%sql

CREATE VIEW sars_view AS SELECT * FROM sars
```

```python
%%sql 

SELECT Date,
       Country,
       [Cumulative number of case(s)] as sum_deaths,
       [Number of deaths] as no_deaths,
       [Number recovered] as no_recovered
       
       
FROM sars_view LIMIT 3
```

```python
%%sql 

SELECT SUM[Number of deaths], Country from sars
GROUP by Country HAVING [Cumulative number of case(s)] > 0 order by [Cumulative number of case(s)] DESC LIMIT 5
```
